# **Key**

```python
goverment_politic_key = ["goverment", "politic", "election","political","parties", "political party"]
health_key = ["covid-19", "corona", "health", "covid", "vaccine", "medicine", "aids", "hiv", "virus"]
economic_key = ["economy", "economic", "recession", "inflation", "economical"]
infrastucture_development_key = ["infrastructure", "facilities", "train", "development","construction"]
crime_key = ["crime", "robbery", "theft", "sexual harassement","illegal", "drug","criminal" "law","criminal"]
tech_industry_key = ["tech", "technology", "industry", "startup", "start-up", "business"]
```

# **XLSum**

```python
from datasets import load_dataset

dataset = load_dataset("csebuetnlp/xlsum","english")
```

```python
df_train =dataset.get("train")
df_test =dataset.get("test")
df_validation =dataset.get("validation")
```

```python
# sum news with key in text
import re

def get_clasification_len_news(key,df,type_news,col):
  sum_news = 0
  for value in df:
    key_found = False
    for k in key:
      if re.findall(k, value[col]):
        key_found = True
        break 
    if key_found:
      sum_news+=1

  print(f'{type_news} news {sum_news} from {len(df)}')
```

```python
# Goverment & Politic
get_clasification_len_news(goverment_politic_key,df_train,"Goverment & Politic","text")
# Health
get_clasification_len_news(health_key,df_train,"Health","text")
# Economic
get_clasification_len_news(economic_key,df_train,"Economic","text")
# Infrastructure Development
get_clasification_len_news(infrastucture_development_key,df_train,"Infrastructure Development","text")
# Crime
get_clasification_len_news(crime_key,df_train,"Crime","text")
# Technology & Industry
get_clasification_len_news(tech_industry_key,df_train,"Technology & Industry","text")
```

```python
def get_non_clasification_len_news(df, type_df):
  sum_news_non_clasification = 0
  for value in df:
    
    key_found = False
    
    for k in goverment_politic_key:
      if re.findall(k, value["text"]):
        key_found = True
        break
    if key_found:
        continue
    for k in health_key:
      if re.findall(k, value["text"]):
        key_found = True
        break
    if key_found:
        continue
    for k in economic_key:
      if re.findall(k, value["text"]):
        key_found = True
        break
    if key_found:
        continue

    for k in infrastucture_development_key:
      if re.findall(k, value["text"]):
        key_found = True
        break
    if key_found:
        continue

    for k in tech_industry_key:
      if re.findall(k, value["text"]):
        key_found = True
        break
    if key_found:
        continue

    sum_news_non_clasification+=1
  print(f'Sum news non clasification {sum_news_non_clasification} from {len(df)} {type_df}')
```

```python
get_non_clasification_len_news(df_test, "df_test")
get_non_clasification_len_news(df_train, "df_train")
get_non_clasification_len_news(df_validation, "df_validation")
```
